## Moving Average:

```python
start = dt.datetime(2015, 1, 1)
end = dt.datetime(2020, 9, 7)
ticker = 'AAPL'
```

```python
data = web.DataReader(ticker, 'yahoo', start, end)
data
```

```python
data['SMA20'] = data['Adj Close'].rolling(20).mean()
data['SMA120'] = data['Adj Close'].rolling(120).mean()
data.head(30)

# 
```

```python
plt.figure(figsize=(12, 8))
ax1 = plt.subplot2grid((7, 1), (0, 0), rowspan = 4, colspan = 1, title = ticker)
ax2 = plt.subplot2grid((7, 1), (5, 0), rowspan = 2, colspan = 1, sharex = ax1)

ax1.plot(data['Adj Close'], label = 'Price', alpha = 0.75)
ax1.plot(data['SMA20'], label = 'SMA20')
ax1.plot(data['SMA120'], label = 'SMA120')

ax2.bar(data.index, data['Volume'], label = 'Volume')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y, %m'))

ax2.set_xlabel('Date (Year - Month)')

ax1.legend()
ax2.legend(loc = 'upper right')

# when SMA20 goes above SMA120, it signal to buy as it suggests that there is a recent surge in buying pressure, 
# that is, investors are willing to buy it more recently as compared to the long-term trend. By the same token, 
# SMA20 going below SMA120 suggests that the price is decreasing recently as compared to long-term trend so we 
# should sell it if we already hold it or maybe short sell it if we don't already own it. Explain short selling
# to the students.
```

```python
data = data.drop(columns = ['High', 'Low', 'Open', 'Close']) # remove the unneeded columns.
```

```python
data
```

```python
data['price_yesterday'] = data['Adj Close'].shift(1)
data.head()

# the adj. close values have been shifted one day forward. Subtracting price_yesterday from adj. close in the same
# row gives absolute change. Dividing this absolute change by price_yesterday gives net percentage change. 
```

```python
(24.714506 - 24.951866) /  24.951866

# every $1 invested in this stock has lost $0.0095 during this one day.
# we can divide the adj close today by price_yesterday, which gives the today worth of $1 invested in this stock
# yesterday as shown below.
```

```python
24.714506 /  24.951866

# this gives us the today's worth of $1 invested in this stock yersterday.
# subtracting 1 from this would give the following.
```

```python
0.9904872845982742 - 1

# this gives the amount lost by investing $1 in this stock as above.
# so, (current price - previous price) / previous price gives us the net gain/loss per $ of investment. In this
# example, it gives us -0.00951, which can be converted to percentage by multiplying it with 100 which then gives
# -0.951%, thus, every $100 invested lost $0.95. Alternatively, we can divide current price by previous price
# (current price / previous price) which gives us the current worth of $1 invested. In this case it is 0.9904 and
# it means that $1 invested in this stock worths $0.9904 today. This can be converted to percentage by multiplying
# with 100 which gives 99.04% which means $100 invested in this stock worths $99.04 today. When 1 subtracted from
# 0.9904, it gives -0.0095 which is the same figure that we earleir obtained by (current price - previous price) / 
# previous price. 
```

```python
data['change'] = data['Adj Close'] / data['price_yesterday']

# it gives the today worth of $1 dollar invested yesterday since they are based on daily adj close prices.
```

```python
data.head()
```

```python
data['invested_SMA'] = np.where(data['SMA20'] > data['SMA120'], 1, 0)
data.head()
```

```python
data['invested_SMA'].describe
```

```python
data['invested_SMA'] == 1
```

```python
data[data['invested_SMA'] == 1]

# Will produce only that part of the data frame where invested_SMA is equal to 1.
```

```python
sma = data[data['invested_SMA'] == 1]
sma
# create a new data frame, sma, containing only those rows invested_SMA is equal to 1. It shows only those rows
# where we have invested based on the strategy when SMA20 was greater than SMA120.
```

```python
# Buy and hold strategy:
```

```python
data['buy_and_hold'] = np.cumprod(data['change'])
data

# $1 invested would have grown to $4.8 at the end of the period. This strategy just says to buy at the start of the
# period and holds till end of the period without  basing buy decisions on SMA20 being greater than SMA120. This
# strategy certainly has worked better than buying when SMA20 is greater than SMA120.
```

#### Exponential weighted moving average strategy:

```python
data['EWM12'] = data['Adj Close'].ewm(span =12, adjust = False).mean()
data['EWM26'] = data['Adj Close'].ewm(span =26, adjust = False).mean()

# More weight is assigned in weighted moving average such as EWM based on timing or frequency. For instance, more
# recent prices could be assigned more weight as compared to relatively old prices.
```

```python
data
```

```python
data['invested_EWM'] = np.where(data['EWM12'] > data['EWM26'], 1, 0)
data
```

```python
ewm = data[data['invested_EWM'] == 1]
ewm
```

```python
ewm['return'] = np.cumprod(ewm['change'])
ewm
```

```python
#%matplotlib notebook
plt.figure(figsize=(14, 10))

ax1 = plt.subplot2grid((10, 1), (0, 0), rowspan = 4, colspan = 1, title = ticker)
ax2 = plt.subplot2grid((10, 1), (5, 0), rowspan = 2, colspan = 1, sharex = ax1)
ax3 = plt.subplot2grid((10, 1), (8, 0), rowspan = 3, colspan = 1, sharex = ax1)

ax1.plot(data['Adj Close'], label = 'Price')
ax1.plot(data['SMA20'], label = 'SMA20')
ax1.plot(data['SMA120'], label = 'SMA120')
ax1.plot(data['EWM12'], label = 'EWM12')
ax1.plot(data['EWM26'], label = 'EWM26')

ax2.bar(data.index, data['Volume'], label = 'Volume')

ax3.plot(data['buy_and_hold'], label = 'Buy and Hold')
ax3.plot(sma['return'], label = 'SMA')
ax3.plot(ewm['return'], label = 'EWM')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y, %m'))

ax3.set_xlabel('Date (Year - Month)')

ax1.legend(fontsize = 10)
ax2.legend(fontsize = 10, loc = 'upper right')
ax3.legend(fontsize = 8)
plt.savefig('ewm.png')
plt.show()

# from the first graph, we can compare EWM12 with EWM26. When the EWM12 goves above EWM26, it is a buy signal and
# we buy. Similarly, we EWM12 goves below EWM26, it is a sell signal and we sell. Based on EWM strategy of buy and
# sell, we generate the return graphed in the last graph, which shows that a higher return has been generated 
# following EWM strategy as compared to SMA and Buy and Hold strategies.

# EWM strategy generates more timely signals to buy and sell as compared to SMA strategy. Therefore, it generates
# superior returns.
```

#### Multiple stocks (assets):

```python
tickers = ['AAPL', 'KO', 'MSFT', 'TSLA', 'AMZN']
```

```python
for ticker in tickers:
    data = web.DataReader(ticker, 'yahoo', start, end)
    data = data.drop(columns = ['High', 'Low', 'Open', 'Close'])
    data['SMA20'] = data['Adj Close'].rolling(20).mean()
    data['SMA120'] = data['Adj Close'].rolling(120).mean()
    data['EWM12'] = data['Adj Close'].ewm(span = 12, adjust = False).mean()
    data['EWM26'] = data['Adj Close'].ewm(span = 26, adjust = False).mean()
    data['price_yesterday'] = data['Adj Close'].shift(1)
    data['change'] = data['Adj Close'] / data['price_yesterday']
    data['invested_SMA'] = np.where(data['SMA20'] > data['SMA120'], 1, 0)
    data['invested_EWM'] = np.where(data['EWM12'] > data['EWM26'], 1, 0)
    sma = data[data['invested_SMA'] == 1] # only take those rows from dataframe, sma, where the condition is satis-
     # fied and store them in new dataframe, sma. The condition is: data['invested_SMA'] == 1.
    ewm = data[data['invested_EWM'] == 1]
    data['buy_and_hold'] = np.cumprod(data['change'])
    sma['return'] = np.cumprod(sma['change']) # sma['change'] is the same column as data['change'] except that
    # data['change'] has all the values whereas sma['change'] has only the values where the condition was met.
    ewm['return'] = np.cumprod(ewm['change'])
    print(ticker)
    print('Buy and Hold return: ' + str(data['buy_and_hold'][-1])) # two strings concatenated.
    print('SMA return: ' + str(sma['return'][-1]))
    print('EWM return: ' + str(ewm['return'][-1]))
```
